```python
# Importa as bibliotecas
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import random
from statistics import mean, stdev, median
from scipy.stats import friedmanchisquare
from prettytable import PrettyTable
import math
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
#from keras.callbacks import EarlyStopping

# Configura os parâmetros do Matplotlib
%matplotlib inline

plt.rcParams.update({'font.size': 20, 
                     'figure.figsize': (10, 8),
                     'figure.dpi': 90,
                     'savefig.dpi': 300,
                     'axes.grid': True,
                     'axes.axisbelow': True, 
                     'grid.color': 'k', 
                     'grid.linestyle': ':', 
                     'grid.linewidth': 0.5,
                     'grid.alpha': 0.5})


# Garante a reprodutibilidade com o Keras
np.random.seed(1)
tf.compat.v1.set_random_seed(2)

# ----------------------------------------------

# Converte as séries em aprendizado supervisionado
# Código adaptado de Jason Brownlee
# https://machinelearningmastery.com/multivariate-time-series-forecasting-lstms-keras/
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    """
    Frame a time series as a supervised learning dataset.
    Arguments:
        data: Sequence of observations as a list or NumPy array.
        n_in: Number of lag observations as input (X).
        n_out: Number of observations as output (y).
        dropnan: Boolean whether or not to drop rows with NaN values.
    Returns:
        Pandas DataFrame of series framed for supervised learning.
    """
      
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    
    # Sequência de entrada (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    
    # Sequência de predição (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    
    # Coloca tudo junto
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    
    # Remove as linhas com valores NaN
    if dropnan:
        agg.dropna(inplace=True)
    return agg

# ----------------------------------------------

# Carrega os dados
df = pd.read_csv('data/artigo/covid19_atributo_valor.csv', header=0, index_col=0)

# Usar termos relativos
termos_relativos = False

# Carrega os dados
if termos_relativos == True:
    
    # Cria uma coluna, transformando os casos em relativos
    df['Casos Relativos'] = df.groupby('Código do País')['Casos Confirmados'].transform(lambda x: x / x.sum())
    
    # Colunas utilizadas, selecionadas de acordo com o método Wrapper    
    cols = [
        'Casos Relativos', 
        'PIB 2019', 
        'População', 
        'Área (Km/2)', 
        'Taxa de Fertilidade', 
        'Média de Idade', 
        'Total de Leitos Hospitalares', 
        'Temperatura Média (Janeiro - Março)',
    ]

else:
    
    # Remove os casos confirmados < 0
    df = df[df['Casos Confirmados'] > 0].cumsum()

    # Teste somente para o Brasil - 24
    #df = df[df['Código do País'] == 24]

    # Colunas utilizadas, selecionadas de acordo com o método Wrapper  
    cols = [
        'Casos Confirmados', 
        'PIB 2019', 
        'População', 
        'Área (Km/2)', 
        'Taxa de Fertilidade', 
        'Média de Idade', 
        'Total de Leitos Hospitalares', 
        'Temperatura Média (Janeiro - Março)',
    ]
    
    
# Seleciona somente os dados das colunas desejadas
dados = df[cols]

print("\n")
print("-" * 100)
print("Predição dos Casos Confirmados de COVID-19 Utilizando Redes LSTM")
print("-" * 100)
print(f"\nAtributo alvo: {cols[0]}")
print(f"\nFeatures utilizados:\n")

for i,c in enumerate(cols[1:]):
    print(f"{i+1}. {c}")
    
print("\n")

# ----------------------------------------------

# Obtém os valores de cada coluna
values = dados.values

# Normaliza os valores
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)

# Determina o número de colunas que serão utilizadas (features)
N_COLS = len(cols) - 1
n_features = N_COLS + 1

# Porcentagem utilizada para gerar o conjunto de treinamento
per_train = 0.8

# Lista com os números de atrasos (delta t)
n_atrasos = [1, 3, 7, 15, 30, 60]

# ----------------------------------------------

# Controla a exibição do arquivo reestruturado
exibir_colunas = False

print("-" * 100)

print("\nAplicação da LSTM...\n")
      
print(f"Deltas: {n_atrasos}\n")

# Lista para armazenar os valores obtidos para a predição (delta, MSE, inv_yhat)
inv_yhat_list = []

# Lista para armazenar as informações dos gráficos de perda (LOSS)
perdas = []

# Loop para aplicação da LSTM
for delta in n_atrasos:
    
    print("." * 100)
    print("\n")
    print(f"* DELTA: {delta}\n")
    
    # Lista para armazenar os valores do MSE e da predição, para o delta atual
    delta_list = []
    
    # Lista para armazena os valores das perda para o delta atual
    perdas_list = []
    
    print("=> 1. Reestruturando o conjunto de dados para o formato adequado...\n")
    
    # Estrutura no formato de aprendizado supervisionado
    reframed = series_to_supervised(scaled, delta, 1)
    print(f"Dimensões do arquivo reestruturado para o formato de aprendizado supervisionado: {reframed.shape}")
    #print("\n")
    
    if exibir_colunas == True:
        print(f"Colunas totais ({len(list(reframed.columns))}): {list(reframed.columns)}")
        print("\n")
        print("Reframed OLD:\n")
        print(reframed.head())
        #print("\n")
    
    # Recupera uma lista com as colunas que devem ser removidas
    colunas = list(reframed.columns)
    removidas = []

    for i in range(2,len(cols)+1):
        for coluna in colunas:
            if coluna == 'var' + str(i) + '(t)':
                removidas.append(coluna)

    # Remove as colunas que não queremos prever
    reframed.drop(removidas, axis=1, inplace=True)

    print(f"Colunas removidas ({len(removidas)}): {removidas}")
    #print("\n")
    print(f"Colunas utilizadas ({len(list(reframed.columns))}): {list(reframed.columns)}")
    print("\n")
    
    if exibir_colunas == True:
        print("Reframed NEW:\n")
        print(reframed.head())
        #print("\n")
        
    # ------------------------------------------
        
    print("=> 2. Preparando os conjuntos de treinamento e de teste...\n")
    
    # Obtém os valores de cada coluna que foi reestruturada
    values = reframed.values
    
    # Define o tamanho do conjunto de teste -> 80% dos dados
    train_size = int(per_train * len(values))
    
    print(f"Conjunto de treinamento gerado com {per_train * 100}% dos dados")
    
    # Separa os dados de treinamento e de testes
    train = values[:train_size, :]
    test = values[train_size:, :]
    
    # Define o total de amostras
    n_amostras = delta * n_features
    
    # Separa os dados de treinamento e de teste
    train_X, train_y = train[:, :n_amostras], train[:, -n_features]
    test_X, test_y = test[:, :n_amostras], test[:, -n_features]
    print(f"Dimensões do conjunto de treinamento: {train_X.shape}, {len(train_X)}, {train_y.shape}")

    # Reestrutura o input para o formato 3D [samples, timesteps, features]
    train_X = train_X.reshape((train_X.shape[0], delta, n_features))
    test_X = test_X.reshape((test_X.shape[0], delta, n_features))
    print(f"Dimensões dos conjuntos de treinamento e de teste: {train_X.shape}, {train_y.shape}, {test_X.shape}, {test_y.shape}")
    print("\n")

    # ------------------------------------------
    
    print("=> 3. Realizando o treinamento da rede neural...\n")
    
    # Configuração utilizada (1, 2 ou 3)
    configuracao = 3
    
    # # Define os parâmetros para a rede neural
    if configuracao == 1:
        neuronios_entrada = 4
        neuronios_saida = 1
        epocas = 10
        batch = 10
        dropout = 0.2
        otimizador = 'adam'
        estimativa_erro = 'mean_squared_error'
        
    elif configuracao == 2:
        neuronios_entrada = 20
        neuronios_saida = 1
        epocas = 100
        batch = 50
        dropout = 0.2
        otimizador = 'adam'
        estimativa_erro = 'mean_squared_error'  
        
    elif configuracao == 3:
        neuronios_entrada = 50
        neuronios_saida = 1
        epocas = 500
        batch = 72
        dropout = 0.2
        otimizador = 'adam'
        estimativa_erro = 'mean_squared_error'
        
    elif configuracao == 0:
        neuronios_entrada = 4
        neuronios_saida = 1
        epocas = 10
        batch = 10
        dropout = 0.2
        otimizador = 'adam'
        estimativa_erro = 'mean_squared_error'
        
    # ------------------------------------------
    
    # Cria o modelo
    model = Sequential()
    model.add(LSTM(neuronios_entrada, input_shape=(train_X.shape[1], train_X.shape[2])))
    model.add(Dropout(dropout))
    model.add(Dense(neuronios_saida))
    model.compile(loss=estimativa_erro, optimizer=otimizador)

    # Realiza o ajuste
    history = model.fit(train_X, train_y, 
                        epochs=epocas, 
                        batch_size=batch, 
                        validation_data=(test_X, test_y), 
                        #callbacks=[EarlyStopping(monitor='val_loss', patience=10)], 
                        verbose=2, 
                        shuffle=False)
    
    # Armzena as informações das perdas
    perdas_list.append(delta)
    perdas_list.append(history)
    perdas.append(perdas_list)
    
    # ------------------------------------------

    # Exibe o resumo
    print("\nResumo do modelo\n")
    print(f"Configuração utilizada: {configuracao}")
    print(f"Time step (delta): {delta}")
    print(f"Total de features: {n_features}")
    print(f"Tamanho do conjunto de treino: {train_size}")
    print(f"Total de neurônios de entrada: {neuronios_entrada}")
    print(f"Total de neurônios de saída: {neuronios_saida}")
    print(f"Total de épocas (epochs): {epocas}")
    print(f"Tamanho do arquivo de lote (batch size): {batch}")
    print(f"Otimizador: {otimizador}")
    print(f"Estimador de erro: {estimativa_erro}")
    print("\n")
    model.summary()
    print("\n")

    # ------------------------------------------
    
    print("=> 4. Plotagem do histórico e da precisão...\n")
    
    plt.figure(figsize=(8, 6))
    plt.plot(history.history['loss'], label='Treino');
    plt.plot(history.history['val_loss'], label='Teste');
    plt.title(f"Gráfico de Desempenho para Delta: {delta}", fontsize=16, y=1.01);
    plt.ylabel('Perda', fontsize=16);
    plt.xlabel('Época', fontsize=16);
    plt.xticks(fontsize=14);
    plt.yticks(fontsize=14);
    plt.legend(bbox_to_anchor=(1.23,1.02), fontsize=13);
    plt.show();
    
    # ------------------------------------------
    
    print("=> 5. Realizando a predição...\n")
    
    testPredict = model.predict(test_X)
    test_X = test_X.reshape((test_X.shape[0], delta * n_features))
    
    trainPredict = model.predict(train_X)
    train_X = train_X.reshape((train_X.shape[0], delta * n_features))

    # Inverte a escala - previsão
    inv_yhat = np.concatenate((testPredict, test_X[:, -N_COLS:]), axis=1)
    inv_yhat = scaler.inverse_transform(inv_yhat)
    inv_yhat = inv_yhat[:,0]
    
     # Inverte a escala - treino
    inv_treino = np.concatenate((trainPredict, train_X[:, -N_COLS:]), axis=1)
    inv_treino = scaler.inverse_transform(inv_treino)
    inv_treino = inv_treino[:,0]

    # Inverte a escala - valor atual
    test_y = test_y.reshape((len(test_y), 1))
    inv_y = np.concatenate((test_y, test_X[:, -N_COLS:]), axis=1)
    inv_y = scaler.inverse_transform(inv_y)
    inv_y = inv_y[:,0]

    # Calcula o MSE
    MSE = mean_squared_error(inv_y, inv_yhat)
    print(f"\nErro Médio Quadrático (MSE): {MSE:.3f}")
    print("\n") 
    
    # ------------------------------------------
    
    # Armazena o delta na lista
    delta_list.append(delta)
    
    # Armazena o MSE na lista
    delta_list.append(MSE)
    
    # Calcula e armazena a média
    media = mean(inv_yhat)
    delta_list.append(media)
    
    # Calcula e armazena o desvio padrão
    desvio = stdev(inv_yhat)
    delta_list.append(desvio)
    
    # Armazena o INV_YHAT na lista -> predições obtidas pelo modelo
    delta_list.append(inv_yhat)
    
    # Armazena todos os valores obtidos para a predição (delta, MSE, inv_yhat)
    inv_yhat_list.append(delta_list)
    
    # ------------------------------------------
    
    print("=> 6. Plotagem da predição...\n")
    
    print("\nGráfico da predição total\n\n")
    
    # Plotagem da predição total
    plt.figure(figsize=(8, 6))
    plt.title(f"Predição de Casos COVID-19, Delta: {delta}", fontsize=20, y=1.01);
    plt.plot(inv_y, color='k', label="Observação");
    plt.plot(inv_yhat, marker='o', color='r', alpha=0.4, label="Modelo");
    plt.ylabel('Casos Confirmados', fontsize=18);
    plt.xlabel('Time Step', fontsize=18);
    plt.xticks(fontsize=16);
    plt.yticks(fontsize=16);
    plt.legend(bbox_to_anchor=(1.3,1.02), fontsize=13);
    plt.show();   
    
    print("\n")
    
    # ------------------------------------------

    # Define alguns pontos 
    N = 100
    pred = 15
    
    print(f"Gráfico da predição para {N} pontos\n\n")

    # Plotagem da predição para esses pontos
    plt.figure(figsize=(8, 6))
    aa = [x for x in range(N)]
    plt.plot(aa, inv_y[:N], marker='o', color='k', alpha=0.4, label="Observação");
    plt.plot(aa, inv_yhat[:N], color='r', label="Modelo");
    plt.title(f"Predição de Casos COVID-19, Delta: {delta}", fontsize=18, y=1.01);
    plt.ylabel('Casos Confirmados', fontsize=18);
    plt.xlabel('Time Step', fontsize=18);
    plt.xticks(fontsize=16);
    plt.yticks(fontsize=16);
    plt.legend(bbox_to_anchor=(1.3,1.02), fontsize=13);
    plt.show();
    
    print("\n")
    
    # ------------------------------------------

    # Define alguns pontos
    N = N-pred

    observacao = inv_y[0:N,]
    modelo = inv_yhat[N:N + pred,]

    x1 = np.arange(len(observacao))
    x2 = np.arange(len(observacao), len(observacao) + pred)
    
    print(f"Gráfico da predição dos {pred} pontos finais\n\n")

    # Plotagem da predição dos pontos finais
    plt.figure(figsize=(8, 6))
    plt.plot(x1, inv_y[:N], marker='o', color='k', alpha=0.4, label="Observacao");
    plt.plot(inv_yhat[0:N + pred,], linewidth=1, color='r', label='Modelo Total');
    plt.plot(x2, modelo, color='b', linewidth=1, label='Modelo Final');
    plt.title(f"Predição de Casos COVID-19, Delta: {delta}", fontsize=18, y=1.01);
    plt.ylabel('Casos Confirmados', fontsize=18);
    plt.xlabel('Time Step', fontsize=18);
    plt.xticks(fontsize=16);
    plt.yticks(fontsize=16);
    plt.legend(bbox_to_anchor=(1.31,1.02), fontsize=13);
    plt.show();
    
# ----------------------------------------------

print("\n")
print("-" * 100)
print("\n")

# ----------------------------------------------

# Salva as predições
nome_arquivo = 'predicoes_' + str(configuracao) + '.csv'
print(f"=> Salvando as predições... -> {nome_arquivo}")

# Cabeçalho da tabela
linha = 'ID,Delta,Inv_Yhat\n'

# Monta a tabela com o resultado final -> ID, Delta, Inv_Yhat
for i, valor in enumerate(inv_yhat_list):
    s = ','.join(map(str, valor[4]))
    linha += f"{i+1}, {valor[0]}, {s}\n"

# Salva as predições em formato CSV
output = open(nome_arquivo, 'w')
output.write(linha)
output.close()

# ----------------------------------------------

# Exibe uma tabela os valores das estísticas
print(f"\n=> Exibindo as estatísticas para a configuração {configuracao}...\n")

# Cabeçalho da tabela
resultado = PrettyTable(['ID', 'Delta', 'Média', 'Mediana', 'Desvio Padrão', 'MSE'], align='c')

# Monta a tabela com o resultado final
for i, valor in enumerate(inv_yhat_list):
    resultado.add_row([i+1, valor[0], np.mean(valor[4]), median(valor[4]), valor[3], valor[1]])

# Utiliza duas casas decimais para os números reais
resultado.float_format = '.2'

# Alinha à direita
resultado.align = 'r'

# Centraliza o ID e o Delta
resultado.align['ID'] = 'c'
resultado.align['Delta'] = 'c'

# Exibe os valores
print(resultado)

# ----------------------------------------------

# Salva as estatísticas
nome_arquivo = 'estatisticas_' + str(configuracao) + '.csv'
print(f"\n=> Salvando as estatísticas... -> {nome_arquivo}")

# Cabeçalho da tabela
linha = 'ID,Delta,Média Global,Mediana,Desvio Padrão,MSE\n'

# Monta a tabela com o resultado final
for i, valor in enumerate(inv_yhat_list):
    linha += f"{i+1},{valor[0]},{np.mean(valor[4]):.2f},{median(valor[4]):.2f},{valor[3]:.2f},{valor[1]:.2f}\n"
    
# Salva as estatísticas em formato csv
output = open(nome_arquivo, 'w')
output.write(linha)
output.close()

# ----------------------------------------------

# Realiza o Teste de Friedman...
print("\n=> Realizando o Teste de Friedman...\n")

print("Hipótese 0: Os valores da predição para o número de casos confirmados de COVID-19 variam de acordo com o Delta T?\n")

# Separa as amostras -> yhat para cada delta
amostras = []
for i, valor in enumerate(inv_yhat_list):
    amostras.append(valor[4])

# Encontra o tamanho da menor amostra
menor_amostra = 100000
for i in amostras:
    if len(i) < menor_amostra:
        menor_amostra = len(i)

# Gera uma lista com as amostras, considerando o tamanho da menor amostra
amostras = [amostra[0:menor_amostra] for amostra in amostras]

# Aplica o Teste de Friedman
stat, p = friedmanchisquare(*amostras)

print(f"Estatística: {stat:.3f}, p: {p:.3f}\n")

# Interpretação
print("Interpretação do resultado...\n")

alpha = 0.05

if p > alpha:
    print("* Mesma distribuição -> falha ao rejeitar a hipótese H0")
else:
    print("* Distribuições diferentes -> rejeita a hipótese H0")

# ----------------------------------------------

# Salva os valores para o Teste de Friedman
print(f"\n=> Exibindo o teste de Friedman para a configuração {configuracao}...\n")

# Cabeçalho da tabela
resultado = PrettyTable(['ID', 'Delta', 'Média', 'Mediana', 'Desvio Padrão', 'Friedman STAT', 'Friedman P'], align='c')

# Monta a tabela com o resultado final
for i, amostra in enumerate(amostras):
    resultado.add_row([i+1, n_atrasos[i], np.mean(amostra), median(amostra), stdev(amostra), stat, p])

# Utiliza duas casas decimais para os números reais
resultado.float_format = '.2'

# Alinha à direita
resultado.align = 'r'

# Centraliza o ID e o Delta
resultado.align['ID'] = 'c'
resultado.align['Delta'] = 'c'

# Exibe os valores
print(resultado)

# ----------------------------------------------

# Salva o teste de Friedman
nome_arquivo = 'friedman_' + str(configuracao) + '.csv'
print(f"\n=> Salvando o teste de Friedman... -> {nome_arquivo}")

# Cabeçalho da tabela
linha = 'ID,Delta,Média,Mediana,Desvio Padrão,Friedman STAT, Friedman P\n'

# Monta a tabela com o resultado final
for i, amostra in enumerate(amostras):
    linha += f"{i+1},{n_atrasos[i]},{np.mean(amostra):.2f},{median(amostra):.2f},{stdev(amostra):.2f},{stat:.2f},{p:.2f}\n"
    
# Salva as estatísticas em formato csv
output = open(nome_arquivo, 'w')
output.write(linha)
output.close()

# ----------------------------------------------

print(f"\n=> Exibindo o comparativo de desempenho...")

# Plotagem do desempenho
plt.figure(figsize=(12, 14));
plt.suptitle(f'Comparativo de Desempenho - Configuração {configuracao}', fontsize=22, y=1.02);

total = len(perdas)
linhas = math.ceil(total/2)

for i in range(total):
    ax = plt.subplot(linhas, 2, i + 1);   
    ax.plot((perdas[i][1]).history['loss'], label='Treino');
    ax.plot((perdas[i][1]).history['val_loss'], label='Teste');
    ax.set_title(f"Gráfico de Desempenho para Delta: {perdas[i][0]}", fontsize=16, y=1.01);
    ax.set_ylabel('Perda', fontsize=16);
    ax.set_xlabel('Época', fontsize=16);
    ax.tick_params(axis='x', labelsize=12);
    ax.tick_params(axis='y', labelsize=12);
    ax.legend(loc='upper right', fontsize=13);  
    
plt.tight_layout();
plt.savefig(f'imagens/desempenho_configuracao_{configuracao}.png', dpi=300, transparent=True, bbox_inches='tight');
plt.show();

# ----------------------------------------------

# Monta o texto para o Latex...

print("\n=> Texto para o Latex...\n")

nome_arquivo = 'estatisticas_' + str(configuracao) + '.csv'

data = pd.read_csv(nome_arquivo, header=0)

cols = ['Delta', 'MSE', 'Mediana', 'Média Global', 'Desvio Padrão']

valores = data[cols].values

for linha in valores:
    print(f"{linha[0]:.0f} & {linha[1]:.3g} & {linha[2]:.3g} & {linha[3]:.3g} & {linha[4]:.3g} \\\\")

# ----------------------------------------------
    
print("\n")
print("-" * 100)
print("Fim!")
print("-" * 100)

# ----------------------------------------------
```
