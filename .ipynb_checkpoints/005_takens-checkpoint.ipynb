```python
# Conta o número de casos confirmados para cada país
df1.groupby(['País/Região'])['Casos Confirmados'].sum().sort_values(ascending=False)
```

```python
# Evolução dos casos confirmados
data = df1.groupby(['Data']).agg({'Casos Confirmados': np.sum}).sort_values(by ='Data', ascending=True)

data.plot(figsize=(20,10), marker='o', markersize=10, linewidth=2, fontsize=14)
plt.xlabel('Data', fontsize=20);
```

```python
# Define as colunas
colunas = ['PIB 2019', 'População', 'População Urbana', 'Área (Km/2)','Densidade (Km/2)', 'Taxa de Fertilidade', 'Média de Idade', 
           'Taxa de Fumantes', 'Taxa de Mortalidade por Doenças Pulmonares', 'Total de Leitos Hospitalares',
           'Temperatura Média (Janeiro - Março)', 'Taxa de Umidade (Janeiro - Março)', 'Casos Confirmados']

# Calcula a matriz de correlação
correlations = df1[colunas].corr()

# Mapa de cor
cmap = sns.diverging_palette(250, 10, as_cmap=True)
#cmap = 'coolwarm'

plt.figure(figsize=(18, 15))

# Plotagem, utilizando o SEABORN
sns.heatmap(correlations, cmap=cmap, 
            xticklabels=correlations.columns, yticklabels=correlations.columns, 
            #xticklabels=False, yticklabels=correlations.columns, 
            cbar=True,
            annot=True, annot_kws={'size': 13});

# Título
plt.suptitle('Matriz de Correlação', x=0.54, y=1.02, fontsize=30);

# Ajusta os labels
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=14)

# Ajusta o layout
plt.tight_layout();

# Salva a figura
#plt.savefig('imagens/covid_australia.png', dpi=300, transparent=True, bbox_inches='tight');
```

```python
# Exibe os dados do Brasil
# Os dados já estão ordenados
df1[df1['País/Região'] == 'Brazil'][['Data', 'País/Região', 'Casos Confirmados']]
```

```python
# Evolução dos casos confirmados - Brasil
plt.figure(figsize=(10, 5))

# País
pais = 'Brazil'

# Dados -> Brasil
valores = df1[df1['País/Região'] == pais]

# Converte para o formato de data
valores['Data'] = pd.to_datetime(valores['Data'])

plt.plot(valores['Data'], valores['Casos Confirmados'], marker='o', markersize=5, linewidth=1)

# Título
plt.suptitle('Casos Confirmados - ' + pais, y=1.05, fontsize=30);

plt.ylabel('Total de Casos', fontsize=20)

# Ajusta os labels
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=14)


# Ajusta o layout
plt.tight_layout();
```
