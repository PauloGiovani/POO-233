```python
# Função que retorna a quantidade de zeros e NaN de cada coluna
def missing_zero_values_table(df):
    
    # Total de valores zero e valores ausentes
    zeros = (df == 0.00).astype(int).sum(axis=0)
    ausentes = df.isnull().sum()
    porcentagem_zeros_ausentes = 100 * df.isnull().sum() / len(df)
    
    # Cria o DataFrame com esses valores
    zero_ausentes_df = pd.concat([zeros, ausentes, porcentagem_zeros_ausentes], axis=1)
    zero_ausentes_df = zero_ausentes_df.rename(columns = {0: 'Zeros', 1: 'NaN', 2: '% Zeros'})
    zero_ausentes_df['Zeros + NaN'] = zero_ausentes_df['Zeros'] + zero_ausentes_df['NaN']
    zero_ausentes_df['% Zeros e NaN'] = 100 * zero_ausentes_df['Zeros + NaN'] / len(df)
    zero_ausentes_df['Tipo de Dados'] = df.dtypes
    
    # Ordena o resultado
    zero_ausentes_df = zero_ausentes_df[zero_ausentes_df.iloc[:,1] != 0].sort_values('% Zeros', ascending=False).round(1)
    
    # Exibe o resultado
    print ("\nO DataFrame possui " + str(df.shape[1]) + " colunas e " + str(df.shape[0]) + " linhas.\n"      
           "Existem " + str(zero_ausentes_df.shape[0]) + " colunas com zeros e valores do tipo NaN.\n")
    
    # Retorna
    return zero_ausentes_df

# Ajusta a quantidade de casas decimais
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# Chama a função, passando o DataFrame
missing_zero_values_table(covid_df)
```

```python
# Função para exibir os valores das barras horizontais
# https://stackoverflow.com/questions/43214978/seaborn-barplot-displaying-values/51535326#51535326
def show_values_on_bars(axs, h_v="v", space=0.4):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height()
                value = int(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - 0.3
                value = int(p.get_width())
                ax.text(_x, _y, value, ha="left", fontsize=14)

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
```

## 3.5 Verificando outliers com Boxplots

Um outlier pode indicar um erro nos dados (como um erro de digitação ou erro de medição, efeitos sazonais etc.). Nesse caso, ele deve ser corrigido ou removido dos dados antes de calcular estatísticas resumidas ou obter informações dos dados, sem as quais levará a análises incorretas.

```python
# Verificando outliers

# Define as colunas
colunas = ['População Urbana', 'Taxa de Fertilidade', 'Média de Idade', 'Taxa de Fumantes', 
           'Taxa de Mortalidade por Doenças Pulmonares', 'Total de Leitos Hospitalares',
           'Temperatura Média (Janeiro - Março)', 'Taxa de Umidade (Janeiro - Março)']

# Props para o boxplot
flierprops = dict(markerfacecolor='0.75', marker='o', markersize=10, linestyle='none')

# Pega somente os valores das colunas desejadas
valores = covid_df[colunas]

# Plotagem
fig, ax = plt.subplots(figsize=(8, 10));

sns.set(style='whitegrid');
ax = sns.boxplot(data=valores, orient='v', palette='colorblind', flierprops=flierprops);
ax.set_title("Verificação de Outliers", fontsize = 20, y=1.01);
ax.set_xticklabels(colunas, rotation=45, ha='right');
ax.tick_params(axis='x', labelsize=14, rotation=45);
ax.tick_params(axis='y', labelsize=14);

# Ajusta o layout
plt.tight_layout();

# Salva a figura
plt.savefig('imagens/outliers_gerais.png', dpi=300, transparent=True, bbox_inches='tight');
```

## 3.7 Matriz de Correlação

Uma matriz de correlação é uma tabela que mostra o valor do coeficiente de correlação (os coeficientes de correlação são usados nas estatísticas para medir a força de um relacionamento entre duas variáveis.) Entre conjuntos de variáveis. Cada atributo do conjunto de dados é comparado com outros atributos para descobrir o coeficiente de correlação. Essa análise permite ver quais pares têm a correlação mais alta, os pares que são altamente correlacionados representam a mesma variação do conjunto de dados, portanto, podemos analisá-los mais para entender qual atributo entre os pares é mais significativo para a construção do modelo.

Na rede de correlação de todas as variáveis selecionadas, o valor da correlação está entre -1 e +1. Variáveis altamente correlacionadas terão valor de correlação próximo a +1 e variáveis menos correlacionadas terão valor de correlação próximo a -1.

```python
# Define as colunas
colunas = ['PIB 2019', 'População', 'População Urbana', 'Área (Km/2)','Densidade (Km/2)', 'Taxa de Fertilidade', 'Média de Idade', 
           'Taxa de Fumantes', 'Taxa de Mortalidade por Doenças Pulmonares', 'Total de Leitos Hospitalares',
           'Temperatura Média (Janeiro - Março)', 'Taxa de Umidade (Janeiro - Março)', 'Casos Confirmados', 'Fatalidades']

# Exibe a correlação de Pearson de algumas colunas
covid_df[colunas].corr()
```

```python
# Plotagem da matriz de correlação

# Calcula a matriz de correlação
correlations = covid_df[colunas].corr()

# Mapa de cor
cmap = sns.diverging_palette(250, 10, as_cmap=True)

# Plotagem
plt.figure(figsize=(18, 15))

sns.heatmap(correlations, cmap=cmap, 
            xticklabels=correlations.columns, yticklabels=correlations.columns, 
            cbar=True,
            annot=True, annot_kws={'size': 13});

# Título
plt.suptitle('Matriz de Correlação', x=0.54, y=1.02, fontsize=30);

# Ajusta os labels
plt.xticks(rotation=45, ha='right', fontsize=14);
plt.yticks(fontsize=14);

# Ajusta o layout
plt.tight_layout();

# Salva a figura
plt.savefig('imagens/matriz_correlacao.png', dpi=300, transparent=True, bbox_inches='tight');
```
