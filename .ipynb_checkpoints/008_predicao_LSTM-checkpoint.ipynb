```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import random

import tensorflow as tf

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
#from keras.callbacks import EarlyStopping

# Configura os parâmetros do Matplotlib
%matplotlib inline

plt.rcParams.update({'font.size': 20, 
                     'figure.figsize': (10, 8),
                     'figure.dpi': 90,
                     'savefig.dpi': 300,
                     'axes.grid': True,
                     'axes.axisbelow': True, 
                     'grid.color': 'k', 
                     'grid.linestyle': ':', 
                     'grid.linewidth': 0.5,
                     'grid.alpha': 0.5})



# Garante a reprodutibilidade com o Keras
np.random.seed(1)
tf.compat.v1.set_random_seed(2)

# ----------------------------------------------

# Converte as séries em aprendizado supervisionado
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    """
    Frame a time series as a supervised learning dataset.
    Arguments:
        data: Sequence of observations as a list or NumPy array.
        n_in: Number of lag observations as input (X).
        n_out: Number of observations as output (y).
        dropnan: Boolean whether or not to drop rows with NaN values.
    Returns:
        Pandas DataFrame of series framed for supervised learning.
    """
      
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    
    # Sequência de entrada (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    
    # Sequência de predição (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    
    # Coloca tudo junto
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    
    # Remove as linhas com valores NaN
    if dropnan:
        agg.dropna(inplace=True)
    return agg

# ----------------------------------------------

print("\n")
print("Predição dos Casos Confirmados de COVID-19 Utilizando Redes LSTM")
print("\n")


# Carrega os dados
df = pd.read_csv('data/covid19-atributo-valor.csv', header=0, index_col=0)

# Remove os casos confirmados < 0
#df = df[df['Casos Confirmados'] > 0]

# Teste somente para o Brasil - 24
#df = df[df['Código do País'] == 24]

# Colunas desejadas
# Verificar por meio da correlação entre as variáveis
cols = [
    'Casos Confirmados',
    #'Código do País', 
    #'Latitude', 
    #'Longitude', 
    #'PIB 2019', 
    #'População',
    'População Urbana', 
    #'Área (Km/2)', 
    'Densidade (Km/2)',
    #'Taxa de Fertilidade', 
    'Média de Idade', 
    'Taxa de Fumantes',
    #'Taxa de Mortalidade por Doenças Pulmonares',
    #'Total de Leitos Hospitalares', 
    #'Temperatura Média (Janeiro - Março)',
    #'Taxa de Umidade (Janeiro - Março)', 
    #'Casos Confirmados'
]


# Seleciona somente as colunas desejadas
dados = df[cols]

N_COLS = len(cols) - 1

# ----------------------------------------------

# Plotagem do número de casos confirmados
fig, ax = plt.subplots(figsize=(10, 6));

sns.set(style='whitegrid');

casos_confirmados = dados.iloc[:, 0]

# Usa um label a cada 2500
ticks_to_use = casos_confirmados.index[::2500]

ax = sns.lineplot(x=casos_confirmados.index, y=casos_confirmados, color='royalblue', marker='o', alpha=0.85);
ax.set_title("Casos Confirmados: COVID-19", fontsize = 18, y=1.01);
ax.set_xlabel(None);
ax.set_ylabel('Total de Casos', fontsize=16);
ax.tick_params(axis='x', labelsize=14, rotation=45 );
ax.tick_params(axis='y', labelsize=14);
ax.set_xticks(ticks_to_use);

# ----------------------------------------------

values = dados.values

# Normaliza os features
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)

# Especifica o número atrasos e de features que serão utilizados
n_atrasos = 31
n_features = N_COLS + 1

# Estrutura no formato de aprendizado supervisionado
reframed = series_to_supervised(scaled, n_atrasos, 1)
print(f"Dimensões do arquivo reestruturado para o formato de aprendizado supervisionado: {reframed.shape}")
print("\n")

print(f"Colunas totais ({len(list(reframed.columns))}): {list(reframed.columns)}")
print("\n")

print("Reframed OLD:\n")
print(reframed.head())
print("\n")

# Recupera uma lista com as colunas que devem ser removidas
colunas = list(reframed.columns)
removidas = []

for i in range(2,len(cols)+1):
    for coluna in colunas:
        if coluna == 'var' + str(i) + '(t)':
            removidas.append(coluna)


# Remove as colunas que não queremos prever
reframed.drop(removidas, axis=1, inplace=True)

print(f"Colunas removidas ({len(removidas)}): {removidas}")
print("\n")

print(f"Colunas utilizadas ({len(list(reframed.columns))}): {list(reframed.columns)}")
print("\n")

print("Reframed NEW:\n")
print(reframed.head())
print("\n")

# ----------------------------------------------

# Divide em conjunto de treino e de teste
values = reframed.values
train_size = int(0.8 * len(values)) #0.8
train = values[:train_size, :]
test = values[train_size:, :]

# Divide em input and output
n_obs = n_atrasos * n_features
#train_X, train_y = train[:, :-1], train[:, -1]
#test_X, test_y = test[:, :-1], test[:, -1]
train_X, train_y = train[:, :n_obs], train[:, -n_features]
test_X, test_y = test[:, :n_obs], test[:, -n_features]
print(f"Dimensões do conjunto de treinamento: {train_X.shape}, {len(train_X)}, {train_y.shape}")
print("\n")

# Reestrutura o input para o formato 3D [samples, timesteps, features]
#train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
#test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
train_X = train_X.reshape((train_X.shape[0], n_atrasos, n_features))
test_X = test_X.reshape((test_X.shape[0], n_atrasos, n_features))
print(f"Dimensões dos conjuntos de treinamento e de teste: {train_X.shape}, {train_y.shape}, {test_X.shape}, {test_y.shape}")
print("\n")

# ----------------------------------------------

# Define a rede neural
print("Treinamento da rede neural:\n\n")
neuronios = 50
epocas = 30
batch = 72 #1
model = Sequential()
model.add(LSTM(neuronios, input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Ajusta o modelo
history = model.fit(train_X, train_y, epochs=epocas, batch_size=batch, validation_data=(test_X, test_y), 
                    #callbacks=[EarlyStopping(monitor='val_loss', patience=10)], 
                    verbose=2, shuffle=False)

print("\n\nResumo do modelo:\n\n")
print(f"Número de atrasos: {n_atrasos}")
print(f"Número de features: {n_features}")
print(f"Tamanho do conjunto de treino: {train_size}")
print(f"Número de Neurônios: {neuronios}")
print(f"Número de Épocas: {epocas}")
print(f"Tamanho do Batch: {batch}")
print("\n")
model.summary()
print("\n")

# ----------------------------------------------

# Plotagem do histórico
#print("\n\nPlotagem do histórico:")
plt.figure();
plt.plot(history.history['loss'], label='Treino');
plt.plot(history.history['val_loss'], label='Teste');
plt.ylabel('Loss', size=15);
plt.xlabel('Epochs', size=15);
plt.legend(fontsize=13);

# ----------------------------------------------

# Realiza uma predição
yhat = model.predict(test_X)
#test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
test_X = test_X.reshape((test_X.shape[0], n_atrasos*n_features))

# Inverte a escala - previsão
inv_yhat = np.concatenate((yhat, test_X[:, -N_COLS:]), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat)
inv_yhat = inv_yhat[:,0]

# Inverte a escala - valor atual
test_y = test_y.reshape((len(test_y), 1))
inv_y = np.concatenate((test_y, test_X[:, -N_COLS:]), axis=1)
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:,0]

# Calcula os erros
print("Erros:\n")
print(f'Test Mean Absolute Error: {mean_absolute_error(inv_y, inv_yhat):.3f}')
print(f'Test Mean Squared Error: {mean_squared_error(inv_y, inv_yhat):.3f}')
print(f'Test Root Mean Squared Error: {np.sqrt(mean_squared_error(inv_y, inv_yhat)):.3f}')
print("\n")

# ----------------------------------------------

plt.figure()

plt.title("Predição de Casos: COVID-19", fontsize=18, y=1.01);
plt.plot(inv_y, color='k', label="Observação");
plt.plot(inv_yhat, marker='o', color='r', alpha=0.4, label="Modelo");
plt.ylabel('Casos Confirmados', fontsize=15);
plt.xlabel('Time Step', fontsize=15);
#plt.legend(loc='best', fontsize=13);
plt.legend(bbox_to_anchor=(1.23,1.017), fontsize=13);
plt.legend(fontsize=13);

# ----------------------------------------------


plt.figure()

# Total de pontos e da predição
N = 500
pred = 100

aa = [x for x in range(N)]
plt.plot(aa, inv_y[:N], marker='o', color='k', alpha=0.4, label="Observação");
plt.plot(aa, inv_yhat[:N], color='r', label="Modelo");
plt.title("Predição de Casos: COVID-19", fontsize=18, y=1.01);
plt.ylabel('Casos Confirmados', fontsize=15);
plt.xlabel('Time Step', fontsize=15);
#plt.legend(loc='best', fontsize=13);
plt.legend(bbox_to_anchor=(1.23,1.017), fontsize=13);
plt.legend(fontsize=13);

# ----------------------------------------------

plt.figure()

# Total de pontos
N = N-pred

observacao = inv_y[0:N,]
modelo = inv_yhat[N:N + pred,]

x1 = np.arange(len(observacao))
x2 = np.arange(len(observacao), len(observacao) + pred)

plt.plot(x1, inv_y[:N], marker='o', color='k', alpha=0.4, label="Observacao");
plt.plot(inv_yhat[0:N + pred,], linewidth=1, color='r', label='Modelo Total');
plt.plot(x2, modelo, color='b', linewidth=1, label='Modelo Final');
plt.title("Predição de Casos: COVID-19", fontsize=18, y=1.01);
plt.ylabel('Casos Confirmados', fontsize=15);
plt.xlabel('Time Step', fontsize=15);
plt.legend(bbox_to_anchor=(1.23,1.017), fontsize=13);
#plt.legend(loc='best', fontsize=13);

# ----------------------------------------------

print("FIM!\n")
```
