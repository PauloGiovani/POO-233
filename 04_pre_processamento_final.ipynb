## 3. Correção da províncias nulas

```python
# Agrupa os dados, selecionando somente as colunas desejadas e aplicando a transformação necessária
# Remove as colunas: Código do País, Província/Estado, Data do Primeiro Caso e Fatalidades
df2 = df.groupby(['Data', 'País/Região'], as_index=False).agg({
    #'Data': 'unique',
    #'Código do País': 'unique',
    'Latitude': 'mean',
    'Longitude': 'mean',
    'PIB 2019': 'unique',
    'População': 'unique',
    'População Urbana': 'unique',
    'Área (Km/2)': 'unique',
    'Densidade (Km/2)': 'unique',
    'Taxa de Fertilidade': 'mean',
    'Média de Idade': 'mean',
    'Taxa de Fumantes': 'mean',
    'Taxa de Mortalidade por Doenças Pulmonares': 'mean',
    'Total de Leitos Hospitalares': 'min',
    'Temperatura Média (Janeiro - Março)': 'mean',
    'Taxa de Umidade (Janeiro - Março)': 'mean',
    'Casos Confirmados': 'sum'
})
```

```python
# Verificando outliers

# Define as colunas
colunas = ['População Urbana', 'Taxa de Fertilidade', 'Média de Idade', 'Taxa de Fumantes', 
           'Taxa de Mortalidade por Doenças Pulmonares', 'Total de Leitos Hospitalares',
           'Temperatura Média (Janeiro - Março)', 'Taxa de Umidade (Janeiro - Março)']

# Props para o boxplot
flierprops = dict(markerfacecolor='0.75', marker='o', markersize=10, linestyle='none')

# Pega somente os valores das colunas desejadas
valores = df2[colunas]

# Plotagem
fig, ax = plt.subplots(figsize=(12, 6))

sns.set(style='whitegrid');

ax = sns.boxplot(data=valores, orient='v', palette='colorblind', flierprops=flierprops);
ax.set_title("Verificação de Outliers", fontsize = 20, y=1.01);
ax.set_xticklabels(colunas, rotation=45, ha='right');
ax.tick_params(axis='x', labelsize=14, rotation=45);
ax.tick_params(axis='y', labelsize=14);
plt.tight_layout();
```

```python
# Define as colunas
colunas = ['PIB 2019', 'População', 'População Urbana', 'Área (Km/2)','Densidade (Km/2)', 'Taxa de Fertilidade', 'Média de Idade', 
           'Taxa de Fumantes', 'Taxa de Mortalidade por Doenças Pulmonares', 'Total de Leitos Hospitalares',
           'Temperatura Média (Janeiro - Março)', 'Taxa de Umidade (Janeiro - Março)', 'Casos Confirmados']

# Calcula a matriz de correlação
correlations = df2[colunas].corr()

# Mapa de cor
cmap = sns.diverging_palette(250, 10, as_cmap=True)

# Plotagem
plt.figure(figsize=(18, 15))

sns.heatmap(correlations, cmap=cmap, 
            xticklabels=correlations.columns, yticklabels=correlations.columns, 
            cbar=True,
            annot=True, annot_kws={'size': 13});

# Título
plt.suptitle('Matriz de Correlação', x=0.54, y=1.02, fontsize=30);

# Ajusta os labels
plt.xticks(rotation=45, ha='right', fontsize=14);
plt.yticks(fontsize=14);

# Ajusta o layout
plt.tight_layout();
```

# 4. Separa o arquivo - Socioeconômicos e Datas

```python
# Separa somente os países e seus dados socioeconômicos

cols = ['País/Região', 'Latitude', 'Longitude', 'PIB 2019', 'População',
       'População Urbana', 'Área (Km/2)', 'Densidade (Km/2)',
       'Taxa de Fertilidade', 'Média de Idade', 'Taxa de Fumantes',
       'Taxa de Mortalidade por Doenças Pulmonares',
       'Total de Leitos Hospitalares', 'Temperatura Média (Janeiro - Março)',
       'Taxa de Umidade (Janeiro - Março)']

socio_df = df2.drop_duplicates('País/Região')[cols]

socio_df
```

```python
# Converte as datas para o formato de string, para poder pivotear as linhas
#df2['Data'].dt.strftime('%Y-%m-%d')

df2['Data'] = df2['Data'].astype(str)

# Exibe algumas colunas
df2[['Data', 'País/Região', 'Casos Confirmados']]
```

```python
# Pivoteia os valores das datas e seus casos confirmados

datas_df = df2.pivot(index='País/Região', columns='Data', values=['Casos Confirmados']).reset_index()

datas_df.columns.name = None

datas_df
```

```python
# Renomeia as colunas

ind = datas_df.columns[1:].map(''.join).str.strip('Casos Confirmados')

ind = ind.insert(0, 'País/Região')
```

```python
ind
```

```python
# Atribui os novos nomes das colunas
datas_df.columns = ind
```

```python
datas_df
```

# 5. Junta tudo em um único arquivo

```python
# Gera um novo DataFrame, unindo as colunas dos DataFrames 'socio' e 'datas'                      
dados = pd.merge(socio_df, datas_df, how ='left', on='País/Região')
```

# 6. Gera o arquivo do tipo atributo-valor

```python
# Cria uma cópia do arquivo de dados
atributo_valor_df = df2.copy(deep=True)

# Converte o atributo País/Região para valores categóricos
atributo_valor_df['País/Região'] = atributo_valor_df['País/Região'].astype('category')

# Gera uma nova coluna com o código do país, começando em 1
atributo_valor_df['Código do País'] = atributo_valor_df['País/Região'].cat.codes + 1

# Define uma nova ordem para as colunas
colunas = [
    'Data',
    #'País/Região',
    'Código do País',
    'Latitude',
    'Longitude',
    'PIB 2019',
    'População',
    'População Urbana',
    'Área (Km/2)',
    'Densidade (Km/2)',
    'Taxa de Fertilidade',
    'Média de Idade',
    'Taxa de Fumantes',
    'Taxa de Mortalidade por Doenças Pulmonares',
    'Total de Leitos Hospitalares',
    'Temperatura Média (Janeiro - Março)',
    'Taxa de Umidade (Janeiro - Março)',
    'Casos Confirmados'
]

# Altera a ordem das colunas
atributo_valor_df = atributo_valor_df[colunas]

# Converte as datas para o formato correto
atributo_valor_df['Data'] = pd.to_datetime(atributo_valor_df['Data'], format='%Y-%m-%d')
```
