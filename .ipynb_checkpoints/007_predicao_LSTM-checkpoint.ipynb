```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import *
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from keras.callbacks import EarlyStopping
import tensorflow as tf
import random
import os

# Configura os parâmetros do Matplotlib
%matplotlib inline

plt.rcParams.update({'font.size': 20, 
                     'figure.figsize': (10, 8),
                     'figure.dpi': 90,
                     'savefig.dpi': 300,
                     'axes.grid': True,
                     'axes.axisbelow': True, 
                     'grid.color': 'k', 
                     'grid.linestyle': ':', 
                     'grid.linewidth': 0.5,
                     'grid.alpha': 0.5})

# ----------------------------------------------

seed_value = 42
print(f"\nSemente para reprodução do experimento: {seed_value}\n")

# Garante a reprodução do experimento
def fix_seeds(seed):
    os.environ['PYTHONHASHSEED'] = '0'
    np.random.seed(seed_value)
    random.seed(seed_value)
    session_conf = tf.compat.v1.ConfigProto(intra_op_parallelism_threads=1, inter_op_parallelism_threads=1) 
    import keras.backend as K
    tf.random.set_seed(seed_value)   
    sess = tf.compat.v1.Session(graph=tf.compat.v1.get_default_graph(), config=session_conf)
    tf.compat.v1.keras.backend.set_session(sess)
    
    #os.environ['PYTHONHASHSEED']=str(2017)
    #tf.random.set_seed(2017)
    #np.random.seed(2017)
    #random.seed(2017)
    
# Não funciona funcionou no TF 2
#fix_seeds(seed_value)

# ----------------------------------------------


# Carrega os dados
#df = pd.read_csv('data/covid19-corrigido.csv', header=0, index_col=0)
df = pd.read_csv('data/covid19-atributo-valor.csv', header=0, index_col=0)

# Remove os casos confirmados < 0
#df = df[df['Casos Confirmados'] > 0]

# Baseado em Apple -> 6 colunas, fora a data
# Colunas utilizadas
# Test Mean Absolute Error: 2863.988
# Test Root Mean Squared Error: 3408.530
cols = [
    'Casos Confirmados',
    #'Código do País', 
    'Latitude', 
    'Longitude', 
    'PIB 2019', 
    'População',
    'População Urbana', 
    #'Área (Km/2)', 
    #'Densidade (Km/2)',
    #'Taxa de Fertilidade', 
    #'Média de Idade', 
    #'Taxa de Fumantes',
    #'Taxa de Mortalidade por Doenças Pulmonares',
    #'Total de Leitos Hospitalares', 
    #'Temperatura Média (Janeiro - Março)',
    #'Taxa de Umidade (Janeiro - Março)', 
    #'Casos Confirmados'
]

# Testes 1 
# Test Mean Absolute Error: 3057.458
# Test Root Mean Squared Error: 3683.908
cols1 = [
    'Casos Confirmados',
    #'Código do País', 
    'Latitude', 
    'Longitude', 
    #'PIB 2019', 
    #'População',
    #'População Urbana', 
    #'Área (Km/2)', 
    #'Densidade (Km/2)',
    #'Taxa de Fertilidade', 
    #'Média de Idade', 
    #'Taxa de Fumantes',
    #'Taxa de Mortalidade por Doenças Pulmonares',
    'Total de Leitos Hospitalares', 
    'Temperatura Média (Janeiro - Março)',
    'Taxa de Umidade (Janeiro - Março)', 
    #'Casos Confirmados'
]

# Testes 2
# Test Mean Absolute Error: 
# Test Root Mean Squared Error: 
cols2 = [
    'Casos Confirmados',
    #'Código do País', 
    #'Latitude', 
    #'Longitude', 
    'PIB 2019', 
    'População',
    'População Urbana', 
    #'Área (Km/2)', 
    'Densidade (Km/2)',
    #'Taxa de Fertilidade', 
    'Média de Idade', 
    'Taxa de Fumantes',
    'Taxa de Mortalidade por Doenças Pulmonares',
    'Total de Leitos Hospitalares', 
    'Temperatura Média (Janeiro - Março)',
    'Taxa de Umidade (Janeiro - Março)', 
    #'Casos Confirmados'
]

# Testes 3
# Test Mean Absolute Error: 903.721
# Test Root Mean Squared Error: 2079.841
# reframed.drop(reframed.columns[[15, 16, 17, 18, 19, 20]], axis=1, inplace=True)
cols3 = [
    'Casos Confirmados',
    #'Código do País', 
    #'Latitude', 
    #'Longitude', 
    #'PIB 2019', 
    'População',
    #'População Urbana', 
    #'Área (Km/2)', 
    'Densidade (Km/2)',
    #'Taxa de Fertilidade', 
    'Média de Idade', 
    'Taxa de Fumantes',
    'Taxa de Mortalidade por Doenças Pulmonares',
    #'Total de Leitos Hospitalares', 
    'Temperatura Média (Janeiro - Março)',
    #'Taxa de Umidade (Janeiro - Março)', 
    #'Casos Confirmados'
]

# Testes 4
# Test Mean Absolute Error: 2096.480
# Test Root Mean Squared Error: 7261.256
#reframed.drop(reframed.columns[[9, 10, 11]], axis=1, inplace=True)
cols = [
    'Casos Confirmados',
    #'Código do País', 
    #'Latitude', 
    #'Longitude', 
    #'PIB 2019', 
    'População',
    #'População Urbana', 
    #'Área (Km/2)', 
    'Densidade (Km/2)',
    #'Taxa de Fertilidade', 
    #'Média de Idade', 
    #'Taxa de Fumantes',
    'Taxa de Mortalidade por Doenças Pulmonares',
    #'Total de Leitos Hospitalares', 
    #'Temperatura Média (Janeiro - Março)',
    #'Taxa de Umidade (Janeiro - Março)', 
    #'Casos Confirmados'
]

# Seleciona somente as colunas desejadas
dados = df[cols]

N_COLS = len(cols) - 1

# ----------------------------------------------

# Plotagem do número de casos confirmados

casos_confirmados = dados.iloc[:, 0]

#casos_confirmados = casos_confirmados[0:15000]

# Usa um label a cada 2500
ticks_to_use = casos_confirmados.index[::2500]

sns.set(style='whitegrid');

fig, ax = plt.subplots(figsize=(10, 6));

#ax = sns.barplot(x=casos_confirmados.index, y=casos_confirmados, color='royalblue', alpha=0.85);
ax = sns.lineplot(x=casos_confirmados.index, y=casos_confirmados, color='royalblue', marker='o', alpha=0.85);

ax.set_title("Casos Confirmados: COVID-19", fontsize = 18, y=1.01);
#ax.set_xlabel("Total de Registros", fontsize=18);
ax.set_xlabel(None);
ax.set_ylabel('Total de Casos', fontsize=16);
#ax.set_xticklabels(casos_confirmados.index, rotation=45, ha='right');
ax.tick_params(axis='x', labelsize=14, rotation=45 );
ax.tick_params(axis='y', labelsize=14);
ax.set_xticks(ticks_to_use);
#ax.set_xticklabels(labels);


# ----------------------------------------------

# Converte as séries em aprendizado supervisionado
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    """
    Frame a time series as a supervised learning dataset.
    Arguments:
        data: Sequence of observations as a list or NumPy array.
        n_in: Number of lag observations as input (X).
        n_out: Number of observations as output (y).
        dropnan: Boolean whether or not to drop rows with NaN values.
    Returns:
        Pandas DataFrame of series framed for supervised learning.
    """
      
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    
    # Sequência de entrada (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    
    # Sequência de predição (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    
    # Coloca tudo junto
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    
    # Remove as linhas com valores NaN
    if dropnan:
        agg.dropna(inplace=True)
    return agg

# ----------------------------------------------


# Normaliza os features
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(dados)

# Especifica o número de horas de atraso
n_hours = 2
n_features = N_COLS + 1

# Estrutura no formato de aprendizado supervisionado
reframed = series_to_supervised(scaled, n_hours, 1)
print("Dimensões do arquivo reestruturado para o formato de aprendizado supervisionado:\n\n")
print(reframed.shape)
print("\n")

print("Reframed OLD:\n")
print(reframed.head())
print("\n")

# Remove as colunas que não queremos prever
#reframed.drop(reframed.columns[[7, 8, 9, 10, 11]], axis=1, inplace=True)
#reframed.drop(reframed.columns[[13, 14, 15, 16, 17]], axis=1, inplace=True)
#reframed.drop(reframed.columns[[15, 16, 17, 18, 19, 20]], axis=1, inplace=True)
reframed.drop(reframed.columns[[9, 10, 11]], axis=1, inplace=True)

print("Reframed NEW:\n")
print(reframed.head())
print("\n")

# Divide em conjunto de treino e de teste
values = reframed.values
train_size = int(0.8 * len(dados))
train = values[:train_size, :]
test = values[train_size:, :]

# Divide em input and output
n_obs = n_hours * n_features
train_X, train_Y = train[:, :n_obs], train[:, -n_features]

test_X, test_Y = test[:, :n_obs], test[:, -n_features]
print("Dimensões do conjunto de treinamento:\n\n")
print(train_X.shape, len(train_X), train_Y.shape)
print("\n")

# Reestrutura o input para o formato 3D [samples, timesteps, features]
train_X = train_X.reshape((train_X.shape[0], n_hours, n_features))
test_X = test_X.reshape((test_X.shape[0], n_hours, n_features))
print("Dimensões dos conjuntos de treinamento e de teste:\n\n")
print(train_X.shape, train_Y.shape, test_X.shape, test_Y.shape)
print("\n")

# ----------------------------------------------


# Define a rede neural
print("Treinamento da rede neural:\n\n")
neuronios = 50
epocas = 5
model = Sequential()
model.add(LSTM(neuronios, input_shape=(train_X.shape[1], train_X.shape[2])))
#model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Ajusta o modelo
history = model.fit(train_X, train_Y, epochs=epocas, batch_size=72, validation_data=(test_X, test_Y), 
                    callbacks=[EarlyStopping(monitor='val_loss', patience=10)], 
                    verbose=2, shuffle=False)

print("\nResumo do modelo:\n")
model.summary()
print("\n")

# Plotagem do histórico
print("\n\nPlotagem do histórico:")
plt.figure();
plt.plot(history.history['loss'], label='Treino');
plt.plot(history.history['val_loss'], label='Teste');
plt.ylabel('Loss', size=15);
plt.xlabel('Epochs', size=15);
plt.legend(fontsize=13);

# ----------------------------------------------

# Realiza uma predição
yhat = model.predict(test_X)
test_X = test_X.reshape((test_X.shape[0], n_hours*n_features))

# Inverte a escala - previsão
inv_yhat = np.concatenate((yhat, test_X[:, -N_COLS:]), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat)
inv_yhat = inv_yhat[:,0]

# Inverte a escala - valor atual
test_Y = test_Y.reshape((len(test_Y), 1))
inv_y = np.concatenate((test_Y, test_X[:, -N_COLS:]), axis=1)
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:,0]

# Calcula os erros
print("\nErros:\n")
print(f'Test Mean Absolute Error: {mean_absolute_error(inv_y, inv_yhat):.3f}')
print(f'Test Root Mean Squared Error: {np.sqrt(mean_squared_error(inv_y, inv_yhat)):.3f}')
#print('Train Mean Absolute Error:', mean_absolute_error(train_Y[0], train_predict[:,0]))
#print('Train Root Mean Squared Error:',np.sqrt(mean_squared_error(Y_train[0], train_predict[:,0])))
print("\n")

# ----------------------------------------------

plt.figure()

N = 100
pred = 10

atual = inv_y[0:N,]
predicao = yhat[N:N + pred,]

x1 = np.arange(len(atual))
x2 = np.arange(len(atual), len(atual) + pred)

plt.plot(x1, test_Y[:N], marker='o', color='k', alpha=0.4, label="Valores Atuais");
plt.plot(yhat[0:N + pred,], linewidth=1, color='r', label='Predição Total');
plt.plot(x2, predicao, color='b', linewidth=1, label='Predição Final');
plt.ylabel('Casos Confirmados', size=15);
plt.xlabel('Time Step', size=15);
plt.legend(fontsize=13);


# ----------------------------------------------

plt.figure()

N = 110

aa=[x for x in range(N)]
plt.plot(aa, test_Y[:N], marker='o', color='k', alpha=0.4, label="Valores Atuais");
plt.plot(aa, yhat[:N], 'r', label="Predição");
plt.ylabel('Casos Confirmados', size=15);
plt.xlabel('Time Step', size=15);
plt.legend(fontsize=13);
#plt.show()

```
